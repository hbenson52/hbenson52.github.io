```python
##import matplotlib and pandas 
import matplotlib.pyplot as plt
import pandas as pd

##read files
dfOpiod = pd.read_csv('C:\\Users\\Cora\\Documents\\haleigh\\BUAD 5112\\CTBA\\M2\\OpidsVA.csv')
dfPres = pd.read_csv('C:\\Users\\Cora\\Documents\\haleigh\\BUAD 5112\\CTBA\\M2\\presidents.csv')
dfTot = pd.read_csv('C:\\Users\\Cora\\Documents\\haleigh\\BUAD 5112\\CTBA\\M2\\TOTALNSA.csv')
```

```python
##change values of median income to 1,000's
dfOpiod['MedianHouseholdIncome'] = (dfOpiod['MedianHouseholdIncome'] / 1000)

##determine Income/Opiod relationship and defineplot type, x/y axis values, size, and color of dots
##I chose a scatterplot to best determine the relationship between income and opiod overdoses. It easily shows how one relates to the other.
ax = dfOpiod.plot(kind='scatter', x='MedianHouseholdIncome', y='FPOO-Rate', figsize=(8,5), color='darkblue')

##set labels for x & y axes and title
ax.set_xlabel('Median Household Income', fontsize=12)
ax.set_ylabel('Number of Opiod Overdoses', fontsize=12)
ax.set_title('Opiod Overdoses Relative to Income ($Thousands) per 100k residents', fontsize=14)
plt.show()
```

1. Do opioid overdoes tend to be associated with less affluent areas—that is, areas where families have lower incomes?
    * According to the data and how it presents itself within the scatterplot, it shows opiod overdoses are associated with less affluent areas. The trend of the plotted data points show the bulk of the of overdoses happen when income is between 30 and 80 thousand with fewer overdoses when income is between 80 and 120 thousand.

```python
##determine Opiod/Heroin relationship and defineplot type, x/y axis values, size, title, and color of dots
##I chose to use a scatterplot for this question as well because it shows the relationship between Opiod/Heroin overdoses clearly
bx = dfOpiod.plot(kind='scatter', x='FPOO-Rate', y='FFHO-Rate', figsize=(8,5))

##set labels and title for new plot
bx.set_xlabel('Number of Opiod Overdoses')
bx.set_ylabel('Number of Heroin Overdoses')
bx.set_title('Opiod Overdoses Relative to Heroin Overdoses per 100k residents')
plt.show()
```

```python
##determine Income/Opiod relationship and defineplot type, x/y axis values, size, title, and color of dots
##I chose a line graph for this question because it includes time, and line graphs are easily read as a line over time.
dx = dfTot.groupby('DATE')['TOTALNSA'].mean().plot(kind='line', x='DATE', y='TOTALNSA', figsize=(10,7))

##set labels and title for new plot
dx.set_xlabel('Date')
dx.set_ylabel('Total Sales')
dx.set_ylim([0,2000])
dx.set_title('Vehicle Sales in the US')
plt.show()
```
